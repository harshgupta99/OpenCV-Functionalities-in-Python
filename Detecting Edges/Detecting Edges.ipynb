```python
import numpy as np
import cv2
from matplotlib import pyplot as plt 
```

```python
image = cv2.imread('cat.jpeg')
```

```python
# To find the edges, we use the Canny Edge Detection method which involves four steps and can be done using the Canny function in opencv.
edge = cv2.Canny(image,50,100)

plt.subplot(121),plt.imshow(image,cmap = 'gray')
plt.title('Original Image')
plt.subplot(122),plt.imshow(edge,cmap = 'gray')
plt.title('Edge Image')
plt.show()
```
