```python
# Smoothing of images can be done by various ways by passing a low pass filter or kernel to convule over the image.
```

```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

```python
# 1) Averaging : This can be done by either using the blur function or the boxfilter function in opencv.
image = cv2.imread('cat.jpeg')

blur1 = cv2.blur(image,(9,9))

plt.subplot(121),plt.imshow(image),plt.title('Original')
plt.subplot(122),plt.imshow(blur1),plt.title('Blurred Image')
plt.show()
```

```python
# 2) Gaussian Filtering : This can be done by using a function called the Gaussianblur. The function getGausiianKernel can be used to obtain the gaussian kernel.
blur2 = cv2.GaussianBlur(image,(5,5),1,1)

plt.subplot(121),plt.imshow(image),plt.title('Original')
plt.subplot(122),plt.imshow(blur2),plt.title('Blurred Image')
plt.show()
```

```python
# 3) Median Filtering : This can be achieved by using a function called the MedianBlur.
median = cv2.medianBlur(image,17)
plt.subplot(121),plt.imshow(image),plt.title('Original')
plt.subplot(122),plt.imshow(median),plt.title('Median Image')
plt.show()
```

```python
# 4) Bilateral Filtering : This removes noise while preserving edges. The function bilateralFilter can be used.
bil = cv2.bilateralFilter(image,3,60,60)

plt.subplot(121),plt.imshow(image),plt.title('Original')
plt.subplot(122),plt.imshow(bil),plt.title('Bilateral Image')
plt.show()
```

```python
# Sharpening : The kernel has all the negative values. Sharpening increases the details in the image. We use the filter2D function to sharpen the image.

sharpening_kernel = np.array([[-1,-1,-1], 
                              [-1, 9,-1],
                              [-1,-1,-1]])

sharpened = cv2.filter2D(image, -1, sharpening_kernel)

plt.subplot(121),plt.imshow(image),plt.title('Original')
plt.subplot(122),plt.imshow(sharpened),plt.title('Sharpened Image')
plt.show()
```
