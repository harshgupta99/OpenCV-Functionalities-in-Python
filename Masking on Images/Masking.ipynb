```python
# Masking of images
# It is basically applying any kind of filter or kernel over the given image. This helps us in performing certain transformations over the image like smoothening, blurring etc.

# To perform 2D convolution over the image, we use a function called the cv2.fliter2D. 

# Let us perform the averaging function using a 17*17 matrix 
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
image = cv2.imread('cat.jpeg')

#We create the kernel
kernel = np.ones((17,17),np.float32)/289

transformed = cv2.filter2D(image,-1,kernel)

plt.subplot(121),plt.imshow(image),plt.title('Original')
plt.subplot(122),plt.imshow(transformed),plt.title('Averaged Image')
plt.show()
```
