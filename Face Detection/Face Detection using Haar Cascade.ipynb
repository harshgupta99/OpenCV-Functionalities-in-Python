```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

```python
# First we load the required XML classsifiers which have been pre-trained for face,eyes, smile detection.
face_detector = cv2.CascadeClassifier('/anaconda3/lib/python3.7/site-packages/cv2/data/haarcascade_frontalface_alt.xml')

#Now we read the image
image = cv2.imread('photo.jpeg')

# Now we convert this 3D image into a 2D image so that we can apply our classifier over it.
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
# Now we use the detectMultiScale feature in the face detector.
faces = face_detector.detectMultiScale(gray,1.05 , 5)
for (x,y,w,h) in faces:
    image = cv2.rectangle(image,(x,y),(x+w,y+h),(255,0,0),2)
    roi_gray = gray[y:y+h, x:x+w]
    roi_color = image[y:y+h, x:x+w]
```

```python
plt.subplot(121),plt.imshow(image,cmap = 'gray')
plt.title('Face Detection in an Image')
```
