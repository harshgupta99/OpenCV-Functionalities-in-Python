```python
# Morphological Transformations : These are simple functions based on the image shape. There are 7 types of basic morphological transformations.

image = cv2.imread('3.jpg')
```

```python
# 2) Dilation : It is done by using a function dilate in opencv.

dilate = cv2.dilate(image,kernel, iterations = 5)
plt.subplot(121),plt.imshow(image),plt.title('Original Image')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(dilate),plt.title('Dilated Image')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
# 3) Opening: This is erosion followed by dilation.

opening = cv2.morphologyEx(image,cv2.MORPH_OPEN,kernel)
plt.subplot(121),plt.imshow(image),plt.title('Original Image')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(opening),plt.title('Opened Image')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
# 4) Closing : Dilation followed by erosion.

closing = cv2.morphologyEx(image,cv2.MORPH_CLOSE,kernel)
plt.subplot(121),plt.imshow(image),plt.title('Original Image')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(closing),plt.title('Closed Image')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
# 5) Morphological Gradient : Difference between dilation and erosion.

morph_grad = cv2.morphologyEx(image,cv2.MORPH_GRADIENT,kernel)
plt.subplot(121),plt.imshow(image),plt.title('Original Image')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(morph_grad),plt.title('Gradient Image')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
# 6) Top Hat : Difference between the input image and the opening of image.

kernel_tophat = np.ones((50,50))

top_hat = cv2.morphologyEx(image,cv2.MORPH_TOPHAT,kernel_tophat)
plt.subplot(121),plt.imshow(image),plt.title('Original Image')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(top_hat),plt.title('Top Hat Image')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
# 7) Black Hat : Difference between closing image and the input image.

kernel_blackhat = np.ones((99,99))
black_hat = cv2.morphologyEx(image,cv2.MORPH_BLACKHAT,kernel_blackhat)
plt.subplot(121),plt.imshow(image),plt.title('Original Image')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(black_hat),plt.title('Black Hat Image')
plt.xticks([]), plt.yticks([])
plt.show()
```
